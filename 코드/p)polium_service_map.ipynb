```python
import folium as g
import pandas as pd
from openpyxl import load_workbook
from folium.plugins import MarkerCluster 
import numpy as np
```

```python
accomo = pd.read_excel('C:/Users/a5215/day12/accomodation_data.xlsx', engine='openpyxl')
accomo
```

시작 지도 만들기

```python
accomo = accomo.dropna(axis=0)
len(accomo)
accomo_map = g.Map(location=[37.879939, 127.727173], tiles = 'Stamen Terrain',zoom_start=8)               
```

마커들 만들기

```python
def plot(accomo):
    locations = list(zip(accomo.위도, accomo.경도))
    icons=[g.Icon(icon='car', prefix="fa", color="red")for _ in range(len(locations))]
    
    cluster = MarkerCluster(locations = locations, icons = icons)
    
    cluster.add_to(accomo_map)
```

```python
for a in range(len(accomo)):
    try : 
        marker_accomo = g.Marker([accomo.loc[a]['위도'], accomo.loc[a]['경도']], icon = g.Icon(color='red'), tooltip=accomo.loc[a]['숙소'])
        marker_accomo.add_to(accomo_map)
    except Exception as e:
        continue
```

```python
for a in range(len(accomo)):
    try : 
        marker_accomo = g.Marker([accomo.loc[a]['위도'], accomo.loc[a]['경도']], icon = g.Icon(color='blue'), tooltip=accomo.loc[a]['여행지'])
        marker_accomo.add_to(accomo_map)
      
    except Exception as e:
        continue
```

```python
plot(accomo)
```
